# 第19页 HAVING 与 WHERE 区别

| WHERE   | HAVING  |
| ------- | ------- |
| 分组前过滤   | 分组后过滤   |
| 不能用聚集函数 | 可以用聚集函数 |

```sql
SELECT Dept, AVG(Grade)
FROM Student JOIN SC
ON Student.ID = SC.StudentID
GROUP BY Dept
HAVING AVG(Grade) > 85;
```

# 第31页 视图（VIEW）概念
我们来详细解释一下SQL中的视图。

可以把视图理解为一个虚拟表。它本身不存储实际的数据，它的内容是由一个SQL查询语句定义的。当你查询视图的时候，数据库就会执行这个被保存的查询语句，然后把结果返回给你。
## 什么是视图
1. 核心概念：虚拟表
普通表：实实在在地存储在硬盘上，占据物理空间，里面存放着数据。

视图：不直接存储数据，只存储一个查询语句。它就像一个窗口或一个快捷方式，透过它，你可以看到来自一个或多个真实表（或其他视图）的数据。
2. 为什么使用视图？（主要优点）
使用视图主要有以下几个好处：

a. 简化复杂查询
假设你经常需要关联四五张表来生成一个销售报表。每次都要写一遍长长的JOIN语句，不仅麻烦，还容易出错。
这时，你可以创建一个视图：
CREATE VIEW 销售报表 AS
SELECT ... FROM 订单表
JOIN 客户表 ON ...
JOIN 商品表 ON ...
JOIN 销售人员表 ON ...
WHERE ...;
之后，你想看这个报表时，只需要简单地查询这个视图就可以了，就像查询一张表一样简单：
SELECT * FROM 销售报表;

```
视图 = 虚拟表（不存储数据）
```

👉 本质是：

```
一个查询语句的结果
```

---

## 作用

* 简化复杂查询
* 提高安全性
* 提高可读性

---

# 第32页 创建视图

## 语法

```sql
CREATE VIEW 视图名[(列名…)]
AS
查询语句
[WITH CHECK OPTION];
```

---

## 示例：创建成绩视图

```sql
CREATE VIEW v_SC AS
SELECT StudentID, CourseID, Grade
FROM SC;
```

DROP VIEW IF EXISTS v_SC;

# 第34页 WITH CHECK OPTION

## 作用

```
保证通过视图修改数据时，仍然满足视图条件
```

---

## 示例

```sql
CREATE VIEW v_high_score AS
SELECT *
FROM SC
WHERE Grade >= 90
WITH CHECK OPTION;
```

---

## 效果

👉 插入或修改数据时：

```sql
INSERT INTO v_high_score VALUES (...)
```

必须满足：

```
Grade >= 90
```

否则报错 ❌
```sql
INSERT INTO v_high_score (StudentID, CourseID, Grade)
VALUES ('200215121', 'C004', 5);
```

# SQL统计与查询核心要素对比表

| 知识点                                 | 核心作用      | 是否依赖分组    | 使用位置        | 典型用途      | 示例                              |
| ----------------------------------- | --------- | --------- | ----------- | --------- | ------------------------------- |
| 聚集函数（COUNT / SUM / AVG / MAX / MIN） | 对数据进行统计计算 | ❌（可单独使用）  | SELECT      | 求总数、平均值等  | `SELECT COUNT(*) FROM Student;` |
| GROUP BY                            | 将数据分组     | ✅         | FROM 之后     | 按类别统计     | `GROUP BY Dept`                 |
| HAVING                              | 对分组结果进行过滤 | ✅         | GROUP BY 之后 | 筛选统计结果    | `HAVING AVG(Grade)>85`          |
| 多表统计（JOIN）                          | 跨表整合数据    | ❌（但常配合分组） | FROM        | 关联多个表进行统计 | `Student JOIN SC`               |
| ORDER BY                            | 对最终结果排序   | ❌         | 最后执行        | 排名、排序展示   | `ORDER BY Grade DESC`           |
| VIEW（视图）                            | 封装查询逻辑    | ❌         | 独立对象        | 简化复杂查询    | `CREATE VIEW v_SC AS ...`       |

# 第37页 SQL执行流程（核心逻辑图）

## SQL执行顺序（非常重要）

```text
FROM → JOIN → WHERE → GROUP BY → HAVING → SELECT → ORDER BY
```

---

## 可视化流程

```text
① FROM / JOIN
   ↓（获取数据）
② WHERE
   ↓（过滤行）
③ GROUP BY
   ↓（分组）
④ HAVING
   ↓（过滤组）
⑤ SELECT
   ↓（计算结果）
⑥ ORDER BY
   ↓（排序输出）
最终结果
```

---

## 示例（完整结构）

```sql
SELECT Dept, AVG(Grade)
FROM Student
JOIN SC ON Student.ID = SC.StudentID
WHERE Age > 18
GROUP BY Dept
HAVING AVG(Grade) > 85
ORDER BY AVG(Grade) DESC;
```

# 第39页 高频错误总结（考试必考🔥）

---

## ❌ 错误1：WHERE 中使用聚集函数

```sql
SELECT Dept
FROM Student JOIN SC
ON Student.ID = SC.StudentID
WHERE AVG(Grade) > 85;   -- ❌ 错误
```

✅ 正确：

```sql
SELECT Student.Dept, AVG(SC.Grade) AS 平均成绩
FROM Student
JOIN SC ON Student.ID = SC.StudentID
GROUP BY Student.Dept
HAVING AVG(SC.Grade) > 85;
```

---

## ❌ 错误2：GROUP BY 后 SELECT 非分组列

```sql
SELECT Dept, Name   -- ❌
FROM Student
GROUP BY Dept;
```

👉 Name 不确定！
```sql
SELECT Dept, COUNT(*) AS 人数
FROM Student
GROUP BY Dept;
```
---

## ❌ 错误3：忘记 GROUP BY

```sql
SELECT Dept, AVG(Grade)  -- ❌
FROM Student JOIN SC
ON Student.ID = SC.StudentID;
```

👉 Dept 没分组
### 正确做法
```sql
SELECT Student.Dept, AVG(SC.Grade) AS 平均成绩
FROM Student
JOIN SC ON Student.ID = SC.StudentID
GROUP BY Student.Dept;
```
---

## ❌ 错误4：误以为 ORDER BY 在前

```text
ORDER BY 是最后执行 ❗
```
### 正确做法
```sql
SELECT Student.Dept, AVG(SC.Grade) AS 平均成绩
FROM Student
JOIN SC ON Student.ID = SC.StudentID
GROUP BY Student.Dept
ORDER BY AVG(SC.Grade) DESC;
```
---

## ❌ 错误5：HAVING 和 WHERE 混淆

| 错误理解         | 正确 |
| ------------ | -- |
| HAVING 用来筛选行 | ❌  |
| WHERE 用来筛选组  | ❌  |
