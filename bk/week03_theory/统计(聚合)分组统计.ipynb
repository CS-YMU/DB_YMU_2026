# 第一部分 聚集函数基础

## 第1页 聚集函数概念

SQL中的 **聚集函数（Aggregate Function）** 用于：

```
对一组数据进行统计计算
```

常见函数：

| 函数       | 含义   |
| -------- | ---- |
| COUNT()  | 统计数量 |
| SUM()    | 求和   |
| AVG()    | 平均值  |
| MAX()    | 最大值  |
| MIN()    | 最小值  |
| STDDEV() | 标准差  |

# 第四部分 成绩统计

---

# 第9页 选课统计

SQL：

```sql
SELECT
COUNT(*) AS 选课人数,
MAX(Grade) AS 最高分,
MIN(Grade) AS 最低分,
SUM(Grade) AS 总分,
AVG(Grade) AS 平均分
FROM SC;
```

---

# 第10页 各课程成绩统计

SQL：

```sql
SELECT
CourseID,
COUNT(*) AS 选课人数,
MAX(Grade) AS 最高分,
MIN(Grade) AS 最低分,
SUM(Grade) AS 总分,
AVG(Grade) AS 平均分
FROM SC
GROUP BY CourseID;
```

---

# 第11页 各课程平均分

SQL：

```sql
SELECT
CourseID,
AVG(Grade) AS 平均分
FROM SC
GROUP BY CourseID;
```

---

# 第五部分 HAVING 条件统计

---

# 第12页 平均分大于85的课程

SQL：

```sql
SELECT
CourseID,
AVG(Grade) AS 平均分
FROM SC
GROUP BY CourseID
HAVING AVG(Grade) > 85;
```

说明：

```
HAVING 用于分组后的条件
```

---

# 第13页 选课人数大于2的课程

SQL：

```sql
SELECT
CourseID,
COUNT(*) AS 选课人数
FROM SC
GROUP BY CourseID
HAVING COUNT(*) > 2;
```

# 第六部分 多表统计

---

# 第14页 各课程名称选课人数

SQL：

```sql
SELECT
Course.CourseName,
COUNT(*) AS 选课人数
FROM SC
JOIN Course
ON SC.CourseID = Course.CourseID
GROUP BY Course.CourseName;
```

---

# 第15页 各系学生平均成绩

SQL：

```sql
SELECT
Student.Dept,
AVG(SC.Grade) AS 平均成绩
FROM Student
JOIN SC
ON Student.ID = SC.StudentID
GROUP BY Student.Dept;
```

# 第七部分 综合统计

---

# 第16页 各课程成绩统计（含课程名）

SQL：

```sql
SELECT
Course.CourseName,
COUNT(*) AS 选课人数,
MAX(SC.Grade) AS 最高分,
MIN(SC.Grade) AS 最低分,
AVG(SC.Grade) AS 平均分
FROM SC
JOIN Course
ON SC.CourseID = Course.CourseID
GROUP BY Course.CourseName;
```

# 第17页 各系选课人数

# 第18页 GROUP BY语法

SQL结构：

```
SELECT 列名, 聚集函数
FROM 表
GROUP BY 分组列
```

示例：

```sql
SELECT Dept, COUNT(*)
FROM Student
GROUP BY Dept;
```

---

# 第19页 HAVING 与 WHERE 区别

| WHERE   | HAVING  |
| ------- | ------- |
| 分组前过滤   | 分组后过滤   |
| 不能用聚集函数 | 可以用聚集函数 |

示例：

```sql
SELECT Dept, AVG(Grade)
FROM Student JOIN SC
ON Student.ID = SC.StudentID
GROUP BY Dept
HAVING AVG(Grade) > 85;
```
