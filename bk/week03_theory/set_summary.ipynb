* **掌握转换能力：**

### 第3页 运算分类速查表

**将抽象符号表格化：**

| 类别 | 运算名称 | 符号 | SQL 对应 | 维度变化 |
| --- | --- | --- | --- | --- |
| **基本** | 选择 (Selection) | $\sigma$ | `WHERE` | 减少行 |
|  | 投影 (Projection) | $\pi$ | `SELECT` | 减少列 |
|  | 笛卡尔积 | $\times$ | `FROM R, S` | 行列暴增 |
| **集合** | 并/交/差 | $\cup, \cap, -$ | `UNION` 等 | 集合缩扩 |
| **特殊** | 连接 / 除 | $\bowtie, \div$ | `JOIN` / 嵌套 | 关联逻辑 |

## 第二部分：基本关系代数 (重点图解)

### 第5页 选择 (Selection)：行的过滤

* **定义：** 根据条件筛选**行**。
* **公式：** $\sigma_{cond}(R)$
* **图解：** 想象一把**水平切片刀**，只留下符合条件的横行。

---

### 第6页 投影 (Projection)：列的过滤

* **定义：** 选出指定的**列**（会自动去重）。
* **公式：** $\pi_{A,B}(R)$
* **图解：** 想象一把**垂直切片刀**，只留下感兴趣的纵列。

---

### 第7页 笛卡尔积：全排列组合

* **现象：** 若 R 有 $n$ 行，S 有 $m$ 行，结果有 $n \times m$ 行。
* **图解：** 每一个 R 的记录都要和 S 的所有记录“握手”一次。
* **警告：** 实际操作中如果不加连接条件，会造成数据灾难。

---

### 第8页 等值连接 (Join)

* **逻辑：** 笛卡尔积 + 过滤条件。
* **对比：**
* $\times$ 是盲目组合。
* $\bowtie$ 是寻找“共同语言”（如 `Student.ID = SC.StudentID`）。

## 第三部分：集合运算 (维恩图解)

### 第10-15页 集合运算三剑客

通过维恩图（Venn Diagram）让学生秒懂逻辑：

| 运算 | 图示需求 | 语义 |
| --- | --- | --- |
| **并 ($\cup$)** | 两个圆全部涂满 | CS系学生 + IS系学生 (全部) |
| **交 ($\cap$)** | 两个圆重叠部分 | 既选了C1又选了C2的学生 |

## 第四部分：除法运算 (逻辑难点)

### 第16-20页 除法运算：全包含逻辑

* **应用场景：** “查询选修了**全部**课程的学生”。
* **逻辑演示表：**

| 学生 | 课程 | 除以 | 目标课程 | 结果 |
| --- | --- | --- | --- | --- |
| 张三 | C1, C2 | $\div$ | {C1, C2} | $\rightarrow$ 张三 |
| 李四 | C1 |  |  | $\rightarrow$ (被排除) |

* **SQL金句：** “不存在一门课，是该学生没有选的。” (双重否定即肯定)
