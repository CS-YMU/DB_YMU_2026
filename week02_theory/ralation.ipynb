# 投影表达式
$\prod_{ID, Name, Sex \text{ as 性别}, Age + 1 \text{ as 明年年龄}, \text{'学生' as 标志}} (Student)$


````python
import pymysql

# 数据库连接配置
config = {
    'host': 'localhost',
    'user': 'dylan',
    'password': 'P@ssw0rd',
    'database': 'student_info_db',
    'charset': 'utf8mb4',
    'cursorclass': pymysql.cursors.DictCursor
}

def execute_projection():
    """执行投影操作：Π_{ID, Name, Sex as 性别, Age + 1 as 明年年龄, '学生' as 标志} (Student)"""
    try:
        # 连接数据库
        conn = pymysql.connect(**config)
        cursor = conn.cursor()
        
        print("🔍 执行投影查询...\n")
        
        # 执行投影操作的 SQL
        sql = """
            SELECT 
                ID, 
                Name, 
                Sex AS '性别', 
                Age + 1 AS '明年年龄', 
                '学生' AS '标志' 
            FROM Student
        """
        
        cursor.execute(sql)
        results = cursor.fetchall()
        
        # 打印表达式
        print("表达式：")
        print("Π_{ID, Name, Sex as 性别, Age + 1 as 明年年龄, '学生' as 标志} (Student)")
        print()
        
        # 打印结果表格
        print("查询结果：")
        print("-" * 60)
        print(f"{'ID':<12} {'Name':<8} {'性别':<4} {'明年年龄':<6} {'标志':<6}")
        print("-" * 60)
        
        for row in results:
            print(f"{row['ID']:<12} {row['Name']:<8} {row['性别']:<4} {row['明年年龄']:<6} {row['标志']:<6}")
        
        print("-" * 60)
        print(f"总记录数: {len(results)} 条")
        
    except pymysql.Error as e:
        print(f"❌ 数据库错误: {e}")
    finally:
        if 'cursor' in locals():
            cursor.close()
        if 'conn' in locals():
            conn.close()
            print("\n🔌 数据库连接已关闭")

# 主程序
if __name__ == "__main__":
    print("="*60)
    print("🚀 开始执行投影查询")
    print("="*60)
    
    # 执行投影查询
    execute_projection()
    
    print("\n✨ 查询完成！")
````

$$σ_{dept = 'CS'} (Student) 或 σ_{Dept = 'CS'} (Student)$$

# 选择运算 σₙ(R): 从表 R 中选出满足条件 F 的元组（行）

形式化表达：
$$σ_F(R) = \{ t | t \in R \land F(t) \text{ 为真} \}$$

F 是条件表达式：有6种基本形态（会写F是关键）

- **比较**：关系表达式  
  - 例如：`Age > 18`、`Salary = 5000`、`Dept ≠ 'CS'`
  
- **复合条件**：逻辑表达式  
  - 例如：`Age > 18 AND Dept = 'CS'`、`Salary > 5000 OR Age < 20`
  
- **属性值是否在闭区间中**  
  - 例如：`Age BETWEEN 18 AND 22`、`Salary BETWEEN 3000 AND 5000`
  
- **属性值是否在集合中**  
  - 例如：`Dept IN ('CS', 'MA', 'IS')`、`Age IN (18, 19, 20)`
  
- **字符串匹配**  
  - 例如：`Name LIKE '张%'`（姓张的学生）、`Name LIKE '%伟'`（名字以"伟"结尾）'%咚咚%'
  
- **属性值是否为空(null)**  
  - 例如：`Phone IS NULL`、`Email IS NOT NULL`

# 条件 F 形式①：关系表达式（比较表达式）

## 关系表达式

| 运算符 | 含义 | 示例 |
|:---:|:---|:---|
| `=` | 等于 | `Age = 18`, `Sex = '男'` |
| `>` | 大于 | `Age > 18`, `Age > Age`, `1 > 1` |
| `>=` | 大于等于 | `Age >= 18`, `1 >= 1`, `1 >= 2` |
| `<` | 小于 | `Age < 18`, `Age < Age`, `1 < 1` |
| `<=` | 小于等于 | `Age <= 18`, `Age <= Age` |
| `!=` 或 `<>` | 不等于 | `Age <> 18`, `Sex != '男'` |

**字符串**：在单引号或双引号对中的字符序列，例如：`'男'`、`"CS"`、`'李勇'`

---

# 条件 F 形式②：逻辑表达式（复合条件）

逻辑表达式使用逻辑运算符连接多个条件，常用的逻辑运算符有：

| 运算符 | 含义 | 示例 |
|:---:|::---|:---|
| **AND** (∧) | 与（所有条件必须同时满足） | `Age > 18 AND Sex = '男'` |
| **OR** (∨) | 或（满足其中一个条件即可） | `Dept = 'CS' OR Dept = 'IS'` |
| **NOT** (¬) | 非（取反，不满足条件） | `NOT (Dept = 'CS')` |

---

## 查询示例

$$σ_{NOT (Age > 18)}(\text{Student})$$

**方法二：直接使用 ≤ 运算符**
$$σ_{Age ≤ 18}(\text{Student})$$

**对应的 SQL 语句：**
```sql

-- 方法2

## 更多复合条件示例

| 查询需求 | 关系代数表达式 | SQL语句 |
|:---|:---|:---|
| CS系或IS系的学生 | $$σ_{Dept = 'CS' \lor Dept = 'IS'}(\text{Student})$$ | `SELECT * FROM Student WHERE Dept = 'CS' OR Dept = 'IS'` |
| 不是CS系的男生 | $$σ_{NOT (Dept = 'CS') \land Sex = '男'}(\text{Student})$$ | `SELECT * FROM Student WHERE Dept != 'CS' AND Sex = '男'` |

---

## 逻辑运算符优先级（从高到低）

1. **NOT** (¬) - 最高优先级
2. **AND** (∧) - 中等优先级
3. **OR** (∨) - 最低优先级

可以使用**括号**改变运算顺序，例如：
$$σ_{(Dept = 'CS' \land Sex = '女') \lor (Age > 19)}(\text{Student})$$

## 其他常见语言陷阱

| 中文表述 | 容易误解为 | 正确逻辑 |
|:---|:---|:---|
| “CS系和IS系的学生” | `Dept = 'CS' AND Dept = 'IS'` ❌ | `Dept = 'CS' OR Dept = 'IS'` ✅ |
| “18岁和20岁的学生” | `Age = 18 AND Age = 20` ❌ | `Age = 18 OR Age = 20` ✅ |
| “不是CS系和IS系的学生” | `NOT (Dept = 'CS' AND Dept = 'IS')` ❌ | `Dept NOT IN ('CS', 'IS')` ✅ |

## 更多语言陷阱示例

| 中文表述 | 容易误解为 | 正确逻辑 |
|:---|:---|:---|
| “年龄在18和20之间的学生” | `Age = 18 AND Age = 20` ❌ | `Age BETWEEN 18 AND 20` ✅ |
| “CS系的男生和女生” | `(Dept='CS' AND Sex='男') AND (Dept='CS' AND Sex='女')` ❌ | `Dept = 'CS'` ✅ |
| “不是CS系的男生和IS系的女生” | `NOT (Dept='CS' AND Sex='男') AND (Dept='IS' AND Sex='女')` ❌ | `(Dept != 'CS' AND Sex='男') OR (Dept='IS' AND Sex='女')` ✅ |

## 陷阱总结

| 陷阱类型 | 说明 | 示例 |
|:---|:---|:---|
| **“和”字陷阱** | 中文“和”有时表示 OR，有时表示 AND | “CS和IS系” → OR |
| **“不是...和...”陷阱** | 否定词与“和”连用时容易混淆 | “不是CS和IS系” → NOT IN |
| **范围表达陷阱** | “在...和...之间”表示区间，不是具体值 | “18和20之间” → BETWEEN |

## 避免陷阱的原则

1. **理解实际语义**：想清楚到底是要“同时满足”还是“满足其中之一”
2. **测试边界情况**：用具体数据测试查询逻辑
3. **使用括号明确优先级**：`(Dept = 'CS' OR Dept = 'IS') AND Age > 18`
4. **考虑使用 IN 操作符**：`Dept IN ('CS', 'IS')` 比 `Dept = 'CS' OR Dept = 'IS'` 更清晰

## 集合运算符：`IN`

表示属性值是否在指定的集合中

| 运算符 | 含义 | 示例 |
|:---:|:---|:---|
| `IN` | 在集合中 | `Dept IN ('CS', 'IS', 'MA')` |
| `NOT IN` | 不在集合中 | `Dept NOT IN ('CS', 'IS')` |

---

## 查询示例

### 1. 查询：CS、IS系的所有学生

**方法一：使用 IN 运算符**
$$σ_{Dept \ IN \ ('CS', 'IS')}(\text{Student})$$

**方法二：使用复合条件（OR 连接）**
$$σ_{Dept = 'CS' \ \lor \ Dept = 'IS'}(\text{Student})$$

**对应的 SQL 语句：**
```sql
-- 方法1：使用 IN
SELECT * FROM Student WHERE Dept IN ('CS', 'IS');

# 条件 F 形式⑥：字符匹配（转义字符）

## 定义转义字符
```sql
属性值 LIKE '\%' ESCAPE '\'
```

`ESCAPE` 后面给出的字符是**转义字符**：在 LIKE 模式的字符串中，转义字符后面的字符表示其本身（而不是通配符）

---

## 查询示例

### 1. 查询名称中有 “%” 或 “_” 字符的课程

**关系代数表达式：**

$
\sigma_{\text{Name LIKE '\%\%\%' OR Name LIKE '\%\_\%'}}(Course)
$

**SQL语句：**
```sql
SELECT * FROM Course 
WHERE Name LIKE '%\%%' ESCAPE '\'
   OR Name LIKE '%*_%' ESCAPE '*';
```

**说明：**
- 第一个条件：匹配包含 `%` 字符的课程（`\%` 表示字符 `%`，两端的 `%` 是通配符）
- 第二个条件：匹配包含 `_` 字符的课程（`*_` 中 `*` 是转义字符，`_` 表示字符 `_`）

**匹配结果示例：**
- ✅ `折扣率%优惠`（包含 %）
- ✅ `100%纯棉`（包含 %）
- ✅ `C_Programming`（包含 _）
- ✅ `Java_基础`（包含 _）
- ❌ `数据结构`（不包含特殊字符）

---

### 2. 查询名称中没有 “%” 字符的课程

**关系代数表达式：**
$
\sigma_{\text{Name 不包含字符 '\%'}}(Course)
$

**SQL语句：**
```sql
SELECT * FROM Course 
WHERE Name NOT LIKE '%\%%' ESCAPE '\';
```

**说明：**
- 使用 `NOT LIKE` 排除包含 `%` 字符的课程
- `\%` 表示字符 `%` 本身
- 两端的 `%` 是通配符

**匹配结果示例：**
- ✅ `数据结构`
- ✅ `操作系统`
- ✅ `C_Programming`（包含 _，但不包含 %，所以匹配）
- ❌ `100%纯棉`（包含 %，被排除）
- ❌ `折扣率%优惠`（包含 %，被排除）

---

## 转义字符示例总结

| 查询需求 | 转义字符 | LIKE模式 | SQL语句 |
|:---|:---:|:---|:---|
| 包含 `%` 字符 | `\` | `'%\%%'` | `SELECT * FROM Course WHERE Name LIKE '%\%%' ESCAPE '\'` |
| 包含 `_` 字符 | `*` | `'%*_%'` | `SELECT * FROM Course WHERE Name LIKE '%*_%' ESCAPE '*'` |
| 包含 `%` 或 `_` | `\` 和 `*` | 复合条件 | `SELECT * FROM Course WHERE Name LIKE '%\%%' ESCAPE '\' OR Name LIKE '%*_%' ESCAPE '*'` |
| 不包含 `%` 字符 | `\` | `'%\%%'` | `SELECT * FROM Course WHERE Name NOT LIKE '%\%%' ESCAPE '\'` |
